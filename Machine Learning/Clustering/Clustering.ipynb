## Clustering

1) What is unsupervised learning in the context of machine learning  

->  

Unsupervised learning is a type of machine learning where the algorithm is trained on unlabeled data, meaning there are no predefined outputs or categories.  
The goal is for the model to discover hidden structures, groupings, or relationships within the dataset on its own.  
Common unsupervised learning techniques include clustering (e.g., K-Means, DBSCAN) and dimensionality reduction (e.g., PCA).  
It is widely used for tasks like market segmentation, anomaly detection, and pattern recognition.



2) How does K-Means clustering algorithm work  

->  

K-Means is an iterative partition-based clustering algorithm that divides data into K non-overlapping clusters.  
It works as follows:  
1. Choose the number of clusters (K) and randomly initialize K centroids.  
2. Assign each data point to the nearest centroid using a distance metric (usually Euclidean distance).  
3. Recalculate the centroids as the mean of all points in each cluster.  
4. Repeat the process until centroids stabilize or maximum iterations are reached.  
The objective is to minimize inertia, the total within-cluster variance.  



3) Explain the concept of a dendrogram in hierarchical clustering  

->  

A dendrogram is a tree-like diagram that shows the hierarchical relationship between clusters.  
It visualizes how data points are merged step-by-step during hierarchical clustering.  
Each leaf of the tree represents an individual observation, and branches indicate cluster combinations.  
By cutting the dendrogram at a certain level (height), one can decide the optimal number of clusters.  
It provides insight into the structure and similarity of the data.



4) What is the main difference between K-Means and Hierarchical Clustering  

->  

The main difference lies in their approach and flexibility:  
- K-Means is a partitional algorithm that divides data into K fixed clusters, requiring K to be defined in advance.  
- Hierarchical clustering builds a hierarchy of clusters without needing K beforehand.  
In hierarchical clustering, once clusters are merged or split, the decision cannot be reversed.  
K-Means is more efficient for large datasets, while hierarchical clustering is better for visualizing small datasets.



5) What are the advantages of DBSCAN over K-Means  

->  

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages:  
- It can find clusters of arbitrary shapes, unlike K-Means which assumes spherical clusters.  
- It automatically determines the number of clusters based on data density.  
- It effectively identifies outliers and noise.  
- DBSCAN is robust to initialization and can handle datasets with varying densities.  
It is especially useful when the data contains irregular or non-linear cluster boundaries.



6) When would you use Silhouette Score in clustering  

->  

The Silhouette Score is used to evaluate clustering quality.  
It measures how similar an object is to its own cluster compared to other clusters.  
The score ranges from -1 to +1:  
- Close to +1 → well-clustered data.  
- Around 0 → overlapping clusters.  
- Negative → misclassified samples.  
It helps choose the optimal number of clusters and compare performance between clustering algorithms.



7) What are the limitations of Hierarchical Clustering  

->  

- Computationally expensive for large datasets (O(n²) complexity).  
- Irreversible merging decisions once clusters are combined.  
- Sensitive to noise and outliers, which can distort the hierarchy.  
- Hard to scale as the dataset grows.  
However, it is valuable for understanding data structure and exploring potential cluster numbers.



8) Why is feature scaling important in clustering algorithms like K-Means  

->  

K-Means relies on distance calculations, typically Euclidean distance.  
If features are on different scales (e.g., age in years vs. income in dollars), the feature with larger values dominates the clustering result.  
Feature scaling (e.g., using StandardScaler or MinMaxScaler) ensures that all features contribute equally to the distance computation.  
This improves accuracy and prevents biased cluster formation.



9) How does DBSCAN identify noise points  

->  

DBSCAN identifies noise based on data density.  
A point is labeled as a core point if it has at least MinPts neighbors within a radius ε.  
Points within the ε-neighborhood of a core point are border points, while those not belonging to any cluster are noise points.  
Thus, DBSCAN naturally separates dense regions from sparse ones, effectively detecting outliers in the data.



10) Define inertia in the context of K-Means  

->  

Inertia measures the compactness of clusters in K-Means.  
It is calculated as the sum of squared distances between each data point and its assigned cluster centroid:  


$$
\text{Inertia} = \sum_{i=1}^{n} \min_{k}\|x_i - \mu_k\|^2
$$


Lower inertia indicates more compact clusters, but extremely low values may mean overfitting.  
Inertia helps assess the optimal number of clusters when used with the Elbow Method.



11) What is the elbow method in K-Means clustering  

->  

The Elbow Method helps determine the optimal number of clusters (K).  
Steps:  
1. Run K-Means with different K values.  
2. Plot K on the x-axis and inertia on the y-axis.  
3. The point where the inertia starts decreasing slowly (forming an “elbow”) is chosen as the best K.  
This balances model simplicity and cluster accuracy.



12) Describe the concept of "density" in DBSCAN  

->  

In DBSCAN, density refers to the number of data points within a neighborhood of radius ε.  
Dense regions represent clusters, while sparse regions represent noise.  
Clusters are formed where the local density exceeds a threshold defined by MinPts.  
This concept allows DBSCAN to identify clusters of varying shapes and sizes without requiring predefined K.



13) Can hierarchical clustering be used on categorical data  

->  

Yes, hierarchical clustering can be applied to categorical data, but it requires special distance measures such as Hamming distance, Jaccard similarity, or Gower distance.  
However, it’s less straightforward than for numerical data since categorical features lack magnitude or order.  
Using proper encoding and distance metrics ensures meaningful results.



14) What does a negative Silhouette Score indicate  

->  

A negative Silhouette Score means that the data point is closer to another cluster than its assigned one.  
This indicates misclassification or overlapping clusters.  
Negative values suggest that the number of clusters may be incorrect or the data may not be well-suited for the chosen algorithm.



15) Explain the term "linkage criteria" in hierarchical clustering  

->  

Linkage criteria determine how the distance between clusters is calculated when merging them.  
Common linkage methods include:  
- Single linkage: Minimum distance between clusters.  
- Complete linkage: Maximum distance between clusters.  
- Average linkage: Mean distance between all points in clusters.  
- Ward linkage: Minimizes total variance within clusters.  
Different linkages can produce different dendrogram shapes and cluster formations.



16) Why might K-Means clustering perform poorly on data with varying cluster sizes or densities  

->  

K-Means assumes clusters are spherical and equally sized.  
When clusters vary in size, density, or shape, it can misclassify data points.  
Smaller or denser clusters might get absorbed into larger ones, and elongated clusters are split incorrectly.  
In such cases, algorithms like DBSCAN or Gaussian Mixture Models perform better.



17) What are the core parameters in DBSCAN, and how do they influence clustering  

->  

The two main parameters are:  
- ε (Epsilon): Defines the neighborhood radius around a point.  
- MinPts: Minimum number of points required to form a dense region.  
A small ε may split clusters, while a large ε may merge distinct clusters.  
MinPts controls how strict the definition of a cluster is — too small leads to noise; too large merges clusters.



18) How does K-Means++ improve upon standard K-Means initialization  

->  

K-Means++ improves the selection of initial centroids to achieve better and faster convergence.  
Steps:  
1. Choose the first centroid randomly.  
2. Choose remaining centroids with a probability proportional to their distance squared from existing centroids.  
3. Proceed with the normal K-Means process.  
This ensures centroids are well-separated initially, reducing the chances of poor clustering due to bad initialization.



19) What is agglomerative clustering  

->  

Agglomerative clustering is a bottom-up approach in hierarchical clustering.  
Initially, each data point starts as its own cluster.  
In each iteration, the two closest clusters (based on distance and linkage criteria) are merged.  
This process continues until all points form a single cluster or the desired number of clusters is reached.  
It is widely used for exploring hierarchical relationships within data.



20) What makes Silhouette Score a better metric than just inertia for model evaluation  

->  

While inertia measures only cluster compactness, it doesn’t consider how distinct clusters are from each other.  
The Silhouette Score evaluates both cohesion (how close points are in a cluster) and separation (how far apart clusters are).  
It provides a value between -1 and +1, allowing easier comparison between different models and cluster counts.  
Hence, it gives a more comprehensive evaluation of clustering quality than inertia alone.

```python
"""
21) Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering.
Visualize using a scatter plot.

->
"""
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate data
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

# Plot
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=30)
plt.scatter(centers[:,0], centers[:,1], c='red', s=200, marker='X')
plt.title("K-Means with 4 Centers (make_blobs)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
"""
31) Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

->
"""
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Generate overlapping data
X, _ = make_blobs(n_samples=400, centers=3, cluster_std=1.8, random_state=42)

# K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)

# Decision boundary visualization
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, cmap=ListedColormap(['#FFD1DC','#B0E0E6','#98FB98']), alpha=0.6)
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', s=30)
plt.title("K-Means Decision Boundaries (Overlapping Data)")
plt.show()
```

```python
"""
43) Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN side-by-side.

->
"""
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Generate data
X43, _ = make_circles(n_samples=600, factor=0.5, noise=0.05, random_state=42)
X43_scaled = StandardScaler().fit_transform(X43)

# KMeans
kmeans43 = KMeans(n_clusters=2, random_state=42)
k_labels43 = kmeans43.fit_predict(X43_scaled)

# DBSCAN
dbscan43 = DBSCAN(eps=0.3, min_samples=5)
d_labels43 = dbscan43.fit_predict(X43_scaled)

# Plot side-by-side
fig43, axs43 = plt.subplots(1, 2, figsize=(12, 5))
axs43[0].scatter(X43_scaled[:,0], X43_scaled[:,1], c=k_labels43, cmap='viridis', s=30)
axs43[0].set_title("KMeans Clustering")
axs43[1].scatter(X43_scaled[:,0], X43_scaled[:,1], c=d_labels43, cmap='plasma', s=30)
axs43[1].set_title("DBSCAN Clustering")
plt.show()
```

```python
"""
44) Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

->
"""
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

# Load data
iris44 = load_iris()
X44 = iris44.data

# Apply KMeans
kmeans44 = KMeans(n_clusters=3, random_state=42)
labels44 = kmeans44.fit_predict(X44)

# Compute Silhouette Coefficients
silhouette_vals44 = silhouette_samples(X44, labels44)

# Plot silhouette values per sample ordered by cluster
plt.figure(figsize=(9,6))
y_lower = 10
n_clusters44 = 3
for i in range(n_clusters44):
    ith_cluster_silhouette_values = silhouette_vals44[labels44 == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values, alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

plt.title("Silhouette Coefficient per Sample (Iris after KMeans)")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Sample index (clustered)")
plt.axvline(x=silhouette_vals44.mean(), color="red", linestyle="--", label="Average silhouette")
plt.legend()
plt.show()
```

```python
"""
45) Generate synthetic data using make_blobs and apply Agglomerative Clustering with 'average' linkage. Visualize clusters.

->
"""
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Generate data
X45, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.6, random_state=42)

# Apply Agglomerative Clustering
agg45 = AgglomerativeClustering(n_clusters=4, linkage='average')
labels45 = agg45.fit_predict(X45)

# Plot
plt.figure(figsize=(8,6))
plt.scatter(X45[:,0], X45[:,1], c=labels45, cmap='viridis', s=30)
plt.title("Agglomerative Clustering (Average Linkage)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
"""
46) Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

->
"""
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data
wine46 = load_wine()
X46 = StandardScaler().fit_transform(wine46.data)
df46 = pd.DataFrame(X46[:, :4], columns=wine46.feature_names[:4])

# KMeans clustering
kmeans46 = KMeans(n_clusters=3, random_state=42)
df46['Cluster'] = kmeans46.fit_predict(X46)

# Pairplot
sns.pairplot(df46, hue='Cluster', diag_kind='kde', palette='Set2')
plt.suptitle("KMeans on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

```python
"""
47) Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

->
"""
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate data with noise
X47, _ = make_blobs(n_samples=500, centers=3, cluster_std=1.0, random_state=42)
# Add some extreme noise points
rng = np.random.RandomState(42)
noise_points = rng.uniform(low=-10, high=10, size=(20, 2))
X47 = np.vstack([X47, noise_points])

# Apply DBSCAN
dbscan47 = DBSCAN(eps=0.8, min_samples=8)
labels47 = dbscan47.fit_predict(X47)

# Count clusters and noise
n_clusters47 = len(set(labels47)) - (1 if -1 in labels47 else 0)
n_noise47 = np.sum(labels47 == -1)
print(f"Clusters Found: {n_clusters47}")
print(f"Noise Points: {n_noise47}")
```

```python
"""
48) Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

->
"""
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load and scale data
digits48 = load_digits()
X48 = StandardScaler().fit_transform(digits48.data)

# Reduce dimensions with t-SNE (may take time)
tsne48 = TSNE(n_components=2, random_state=42, perplexity=30)
X48_tsne = tsne48.fit_transform(X48)

# Apply Agglomerative Clustering
agg48 = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels48 = agg48.fit_predict(X48_tsne)

# Plot clusters
plt.figure(figsize=(8,6))
plt.scatter(X48_tsne[:,0], X48_tsne[:,1], c=labels48, cmap='tab10', s=20)
plt.title("Agglomerative Clustering on Digits (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
