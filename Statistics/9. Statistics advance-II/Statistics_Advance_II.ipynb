1. What is hypothesis testing in statistics?

-> Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to draw conclusions about a population. It's a formal procedure for investigating our ideas about the world using statistics.




2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

->

The null hypothesis (H₀) is a statement that there is no significant difference between a specified population parameter and a hypothesized value, or no significant difference between two populations. It is the statement that the researcher is trying to disprove.

The alternative hypothesis (H₁) is a statement that contradicts the null hypothesis. It is the statement that the researcher is trying to prove.

It differs from the alternative hypothesis in the following ways:-

    i) Purpose: The null hypothesis is the status quo, while the alternative hypothesis is the claim or idea being tested.

    ii) Symbol: Null hypothesis is denoted by H₀, and the alternative hypothesis is denoted by H₁.

    iii) Goal of testing: The goal of hypothesis testing is to determine if there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.
     
    iv) Equality: The null hypothesis always includes a statement of equality (e.g., =, ≤, or ≥), while the alternative hypothesis never includes a statement of equality (e.g., ≠, >, or <).


3.  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

->
The significance level, denoted by the Greek letter alpha (α), is a crucial concept in hypothesis testing. It represents the maximum probability of making a Type I error that you are willing to accept. In other words, it's the probability of rejecting the null hypothesis when it is actually true.

Here it's role in deciding the outcome of a test:

  i) Setting the Threshold:
  
  Before conducting the hypothesis test, you set the significance level (α). Common values are 0.05 (or 5%), 0.01 (1%), or 0.10 (10%). This value serves as a threshold for your decision.
  
  ii) Comparing with the P-value:
  
  After performing the statistical test and calculating the test statistic, you obtain a p-value. The p-value is the probability of observing your sample data (or more extreme data) if the null hypothesis were true.
  
  iii) Making the Decision:
  
  The decision to reject or fail to reject the null hypothesis is made by comparing the p-value to the significance level:
      i) If p-value ≤ α:
      
      You reject the null hypothesis. This means that the observed data is unlikely to have occurred by random chance if the null hypothesis were true. The evidence from your sample is considered statistically significant at the chosen significance level, providing support for the alternative hypothesis.
      
      ii) If p-value > α:
      
      You fail to reject the null hypothesis. This means that the observed data is reasonably likely to have occurred by random chance even if the null hypothesis were true. The evidence from your sample is not considered statistically significant at the chosen significance level, and you do not have enough evidence to conclude that the alternative hypothesis is true.




4. What are Type I and Type II errors? Give examples of each.

->

  Type I Error (False Positive):

    This occurs when you reject the null hypothesis (H₀) when it is actually true. You conclude there is a significant effect or difference when there isn't one in reality. The probability of making a Type I error is equal to the significance level (α) you set for your test.

    Example:

    A medical test is designed to detect a specific disease. A Type I error would be if the test results are positive, indicating the presence of the disease, but the person is actually healthy.


  Type II Error (False Negative):

    This occurs when you fail to reject the null hypothesis (H₀) when it is actually false. You conclude there is no significant effect or difference when there is one in reality. The probability of making a Type II error is denoted by β (beta).

    Example:

    In the same medical test scenario, a Type II error would be if the test results are negative, indicating the absence of the disease, but the person actually has the disease.




5. What is the difference between a Z-test and a T-test? Explain when to use

->
Both Z-tests and T-tests are statistical hypothesis tests used to compare means. However, they are used under different conditions and rely on different assumptions:

**Z-test:**

* **When to use:** A Z-test is used when the population standard deviation is known, or when the sample size is large (typically n > 30), allowing the sample standard deviation to be a good approximation of the population standard deviation due to the Central Limit Theorem.

* **Distribution:** It uses the standard normal distribution (Z-distribution).

* **Formula:** The Z-statistic is calculated as:

   $Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$

   where:
   - $\bar{x}$ is the sample mean
   - $\mu$ is the population mean
   - $\sigma$ is the population standard deviation
   - $n$ is the sample size

**T-test:**

* **When to use:** A T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30). In this case, the sample standard deviation is used as an estimate of the population standard deviation, which introduces more uncertainty.

* **Distribution:** It uses the t-distribution (with degrees of freedom)  , which has heavier tails than the normal distribution to account for the increased uncertainty with smaller sample sizes. The shape of the t-distribution depends on the degrees of freedom, which are related to the sample size (typically n-1 for a one-sample t-test).

* **Formula:** The T-statistic is calculated as:

   $T = \frac{\bar{x} - \mu}{s / \sqrt{n}}$

   where:
   - $\bar{x}$ is the sample mean
   - $\mu$ is the population mean
   - $s$ is the sample standard deviation
   - $n$ is the sample size


In practice, the t-test is more commonly used because the population standard deviation is rarely known. However, for large sample sizes, the t-distribution approaches the normal distribution, and the results of a Z-test and T-test will be very similar.

6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.  

Hint: Generate random number using random function.

->

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5 # Probability of success

# Generate random numbers from a binomial distribution
binomial_data = np.random.binomial(n, p, size=1000)

# Plot the histogram
plt.hist(binomial_data, bins=np.arange(n+2)-0.5, density=True, edgecolor='black')
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xticks(np.arange(n+1))
plt.show()
```

7.  Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

->

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Calculate the sample mean
sample_mean = np.mean(sample_data)
print(f"Sample Mean: {sample_mean:.2f}")


# Null hypothesis: the population mean is 50
mu = 50


# Perform the Z-test
# The ztest function in statsmodels calculates the standard deviation from the sample data.
# In a real-world scenario with unknown population standard deviation, a t-test would be more appropriate.
z_statistic, p_value = ztest(sample_data, value=mu, alternative='two-sided')

print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
```

8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

->

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters for the normal distribution
mu = 100  # Mean
sigma = 15  # Standard deviation
sample_size = 200 # Sample size

# Simulate data from a normal distribution
data = np.random.normal(mu, sigma, sample_size)

# Calculate the 95% confidence interval for the mean
confidence_interval = stats.norm.interval(0.95, loc=np.mean(data), scale=stats.sem(data))

print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

# Plot the data using a histogram
plt.hist(data, bins=20, density=True, alpha=0.6, color='g')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simulated Normal Distribution Data")
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

->

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_zscores(data):


  # Calculate Z-scores
  z_scores = zscore(data)

  print("Original Data (first 10):", data[:10])
  print(f"Original Data Mean: {np.mean(data):.2f}")
  print(f"Original Data Standard Deviation: {np.std(data):.2f}")
  print("Z-scores (first 10):", z_scores[:10])
  print(f"Z-scores Mean: {np.mean(z_scores):.2f}")
  print(f"Z-scores Standard Deviation: {np.std(z_scores):.2f}")


  # Plot the histogram of standardized data
  plt.hist(z_scores, bins=20, density=True, alpha=0.6, color='purple')
  plt.xlabel("Z-score")
  plt.ylabel("Frequency")
  plt.title("Histogram of Standardized Data (Z-scores)")
  plt.show()

# Define a completely different dataset
new_dataset = [15, 22, 28, 35, 40, 48, 55, 62, 68, 75, 80, 88, 95, 102, 110]

calculate_and_plot_zscores(new_dataset)
```
