1) What is a Convolutional Neural Network (CNN), and how does it differ from traditional fully connected neural networks in terms of architecture and performance on image data?

->


Convolutional Neural Network (CNN)  
A CNN is a type of neural network designed specifically for image data. It uses convolutional filters to automatically detect features like edges, textures, and shapes.

How CNNs differ from Fully Connected Networks:

- Architecture:
  - CNNs use local connections and shared weights through convolutional layers.
  - Fully connected networks connect every neuron to all neurons in the next layer.
  - CNNs preserve the spatial structure of images; fully connected networks do not.

- Performance on Image Data:
  - CNNs have fewer parameters, making them more efficient and less prone to overfitting.
  - They learn hierarchical features (edges → shapes → objects) and generalize better.
  - CNNs achieve much higher accuracy on image tasks compared to fully connected models.



2) Discuss the architecture of LeNet-5 and explain how it laid the foundation for modern deep learning models in computer vision. Include references to its original research paper.

->

Architecture of LeNet-5

LeNet-5, proposed by Yann LeCun et al. (1998) in the paper *"Gradient-Based Learning Applied to Document Recognition"*, is one of the earliest and most influential convolutional neural network architectures. It was originally designed for handwritten digit recognition (e.g., MNIST).

LeNet-5 Architecture Overview:

1. Input Layer
   - 32×32 grayscale image.

2. C1 — Convolutional Layer
   - 6 feature maps
   - 5×5 filters
   - Outputs 28×28 feature maps (no padding)

3. S2 — Subsampling (Pooling) Layer
   - Average pooling
   - 2×2 filter with stride 2
   - Outputs 14×14 feature maps

4. C3 — Convolutional Layer
   - 16 feature maps
   - 5×5 filters
   - Outputs 10×10 feature maps

5. S4 — Subsampling Layer
   - Average pooling
   - 2×2 filter with stride 2
   - Outputs 5×5 feature maps

6. C5 — Convolutional Layer (Fully Connected to Previous Maps)
   - 120 feature maps
   - 5×5 filters (acts like fully connected layer)

7. F6 — Fully Connected Layer
   - 84 neurons

8. Output Layer
   - 10-way softmax (for digit classification)


How LeNet-5 Laid the Foundation for Modern Computer Vision Models

- Introduced core CNN concepts
  - Convolution layers
  - Subsampling (pooling)
  - Feature extraction hierarchy

- Demonstrated weight sharing
  - Reduced the number of parameters dramatically compared to fully connected networks.
  - Inspired future architectures like AlexNet, VGG, and ResNet.

- Showed the power of deep feature learning
  - Learned features automatically from raw pixels.
  - Provided an alternative to hand-crafted features (e.g., SIFT, HOG).

- Pioneered end-to-end training
  - Entire pipeline trained with backpropagation, influencing all modern deep learning systems.

- Inspired scalable deep CNNs
  - Although shallow by today’s standards, its principles scale directly to deeper networks (e.g., convolution–pooling stacks).






3) Compare and contrast AlexNet and VGGNet in terms of design principles, number of parameters, and performance. Highlight key innovations and limitations of each.

->


Comparison of AlexNet and VGGNet

 1. Design Principles

AlexNet (2012)  
- Introduced deeper CNNs to large-scale image classification (ImageNet).  
- Used ReLU activation, which accelerated training compared to tanh/sigmoid.  
- Employed dropout to reduce overfitting.  
- Used overlapping max-pooling and local response normalization (LRN).  
- Trained using two GPUs due to limited hardware.

VGGNet (2014)  
- Emphasized architectural simplicity and uniformity.  
- Used stacks of 3×3 convolutions with very deep networks (VGG-16, VGG-19).  
- Removed LRN and relied on depth and small filters for performance gains.  
- Focused on showing that depth alone improves accuracy.


 2. Number of Parameters

| Model        | Depth | Parameters |
|--------------|-------|------------|
| AlexNet  | 8 layers | ~60 million |
| VGG-16   | 16 layers | ~138 million |
| VGG-19   | 19 layers | ~144 million |

- AlexNet is large but manageable.  
- VGGNet has more than double the parameters, making it memory-heavy and computationally expensive.



 3. Performance (ImageNet)

- AlexNet: Top-5 error ≈ 15.3%  
- VGGNet (VGG-16): Top-5 error ≈ 7.3%  

VGGNet significantly outperformed AlexNet due to greater depth and more expressive features.



 4. Key Innovations

AlexNet Innovations  
- ReLU activation for faster training.  
- Dropout for regularization.  
- GPU-based training.  
- Overlapping max-pooling.  
- First breakthrough deep CNN in the ImageNet challenge.


VGGNet Innovations  
- Demonstrated the power of very deep networks (up to 19 layers).  
- Used small 3×3 filters, which became a standard for CNN design.  
- Simple and modular architecture inspired many later models (e.g., ResNet).



 5. Limitations

AlexNet Limitations  
- Relatively shallow compared to modern networks.  
- LRN did not provide long-term benefits and was eventually dropped.  
- Architecture less clean and uniform.


VGGNet Limitations  
- Extremely large number of parameters → high memory usage.  
- Slow to train and deploy.  
- Not optimized for computational efficiency.


Both architectures shaped modern deep learning, but VGGNet’s simplicity and AlexNet’s innovations each played pivotal roles in developing today’s CNNs.





4) What is transfer learning in the context of image classification? Explain how it helps in reducing computational costs and improving model performance with limited data.

->

Transfer Learning in Image Classification

Transfer learning is a technique where a model pre-trained on a large dataset (such as ImageNet) is reused for a new, typically smaller, image classification task. Instead of training a model from scratch, you start with a network that already learned useful visual features.



 How Transfer Learning Works
- A CNN pre-trained on millions of images learns general features like edges, textures, and shapes.
- These learned features are reused by:
  - Fine-tuning: adjusting some or all pre-trained layers on the new dataset.
  - Feature extraction: freezing the pre-trained layers and training only the final classifier.


 Benefits

 1. Reduces Computational Cost
- Training deep CNNs from scratch requires massive compute.
- Using a pre-trained model avoids training all layers, saving time and GPU resources.

 2. Improves Performance with Limited Data
- Small datasets risk overfitting when training from scratch.
- Pre-trained models provide strong, generalized features that improve accuracy.
- Fine-tuning helps adapt these features to the target domain.



Transfer learning allows models to leverage previously learned knowledge, enabling faster, cheaper, and more accurate training—especially when data is limited.





5) Describe the role of residual connections in ResNet architecture. How do they address the vanishing gradient problem in deep CNNs?  

->

Residual Connections in ResNet

Residual connections are skip connections that add the input x directly to the output of a block F(x):

`output = F(x) + x`



 Role of Residual Connections
- Allow layers to learn residuals (small modifications) instead of full transformations.
- Help preserve important information by passing it forward unchanged.
- Enable very deep networks (50, 101, 152+ layers) to train effectively.



 How They Address the Vanishing Gradient Problem
- Skip connections create a direct path for gradients to flow backward.
- Gradients don’t have to pass through many nonlinear layers, preventing them from shrinking.
- This stabilizes training and ensures effective learning even in extremely deep CNNs.



 In Summary
Residual connections improve gradient flow, reduce vanishing gradients, and make deep architectures practical and efficient.

```python
"""
6) Implement the LeNet-5 architectures using Tensorflow or PyTorch to classify the MNIST dataset. Report the accuracy and training time.

->

"""

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import time


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
])

train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_dataset  = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader  = DataLoader(test_dataset, batch_size=1000, shuffle=False)



class LeNet5(nn.Module):
    def __init__(self):
        super(LeNet5, self).__init__()
        self.conv1 = nn.Conv2d(1, 6, kernel_size=5)       # C1
        self.pool = nn.AvgPool2d(2, stride=2)             # S2, S4
        self.conv2 = nn.Conv2d(6, 16, kernel_size=5)      # C3
        self.conv3 = nn.Conv2d(16, 120, kernel_size=5)    # C5
        self.fc1 = nn.Linear(120, 84)                     # F6
        self.fc2 = nn.Linear(84, 10)                      # Output


    def forward(self, x):
        x = F.tanh(self.conv1(x))
        x = self.pool(x)
        x = F.tanh(self.conv2(x))
        x = self.pool(x)
        x = F.tanh(self.conv3(x))
        x = x.view(-1, 120)
        x = F.tanh(self.fc1(x))
        x = self.fc2(x)
        return x


model = LeNet5().to(device)



criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)




epochs = 5
start_time = time.time()

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_loader):.4f}")

training_time = time.time() - start_time




model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = correct / total * 100

print(f"\nFinal Test Accuracy: {accuracy:.2f}%")
print(f"Training Time: {training_time:.2f} seconds")
```

```python
"""
7)  Use a pre-trained VGG16 model (via transfer learning) on a small custom dataset (e.g., flowers or animals).

Replace the top layers and fine-tune the model. Include your code and result discussion.

->

"""
# !pip install tensorflow-datasets

import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt


IMG_SIZE = 128   # Reduced from 224
BATCH_SIZE = 16  # Smaller batch = faster epochs
EPOCHS_STAGE1 = 2
EPOCHS_STAGE2 = 1


(ds_train, ds_val), ds_info = tfds.load(
    "tf_flowers",
    split=["train[:80%]", "train[80%:]"],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)

NUM_CLASSES = ds_info.features["label"].num_classes


def preprocess(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image / 255.0
    return image, label

augment = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
])

train = (
    ds_train
    .map(preprocess)
    .map(lambda x, y: (augment(x), y))
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)

val = (
    ds_val
    .map(preprocess)
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)


base_model = VGG16(
    include_top=False,
    weights="imagenet",
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

for layer in base_model.layers:
    layer.trainable = False


x = Flatten()(base_model.output)
x = Dense(128, activation='relu')(x)   # smaller dense layer
x = Dropout(0.4)(x)
output = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    optimizer=Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)


history1 = model.fit(
    train,
    validation_data=val,
    epochs=EPOCHS_STAGE1
)


for layer in base_model.layers[-3:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)


history2 = model.fit(
    train,
    validation_data=val,
    epochs=EPOCHS_STAGE2
)

acc = history1.history["accuracy"] + history2.history["accuracy"]
val_acc = history1.history["val_accuracy"] + history2.history["val_accuracy"]

plt.plot(acc, label="Train Accuracy")
plt.plot(val_acc, label="Val Accuracy")
plt.legend()
plt.title("VGG16 Transfer Learning — Fast Version")
plt.show()

loss, accuracy = model.evaluate(val)
print("Final Validation Accuracy:", accuracy)
```

```python
"""
8)  Write a program to visualize the filters and feature maps of the first
convolutional layer of AlexNet on an example input image.

->

"""

import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import requests
from io import BytesIO

# Load pretrained AlexNet
alexnet = models.alexnet(weights='IMAGENET1K_V1')
alexnet.eval()

# First convolutional layer
first_conv = alexnet.features[0]
print(first_conv)

url = "https://images.unsplash.com/photo-1518791841217-8f162f1e1131"  # dog image
response = requests.get(url)
img = Image.open(BytesIO(response.content)).convert("RGB")
img

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

input_tensor = transform(img).unsqueeze(0)  # shape: (1, 3, 224, 224)


filters = first_conv.weight.data.clone()

num_filters = filters.shape[0]  # 64 filters
plt.figure(figsize=(15, 10))

for i in range(32):  # show first 32 filters
    f = filters[i]
    f_min, f_max = f.min(), f.max()
    f = (f - f_min) / (f_max - f_min)  # normalize for visualization

    plt.subplot(4, 8, i+1)
    plt.imshow(f.permute(1, 2, 0))  # CHW → HWC
    plt.axis("off")

plt.suptitle("AlexNet - First Conv Layer Filters", fontsize=18)
plt.show()

with torch.no_grad():
    feature_maps = first_conv(input_tensor)  # shape: (1, 64, H, W)

feature_maps.shape

fm = feature_maps.squeeze(0)  # (64, H, W)

plt.figure(figsize=(15, 10))

for i in range(32):  # show first 32 feature maps
    plt.subplot(4, 8, i+1)
    plt.imshow(fm[i].cpu().numpy(), cmap='gray')
    plt.axis("off")

plt.suptitle("AlexNet - First Conv Layer Feature Maps", fontsize=18)
plt.show()
```
